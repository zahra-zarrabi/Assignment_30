```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
```

```python
image=cv2.imread('rubix.png')
image2=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
h,s,v=cv2.split(image2)
plt.imshow(image,cmap='gray')
```

```python
lower_cyan=np.array([80,35,140])
upper_cyan=np.array([100,255,255])

lower_yellow=np.array([20,35,140])
upper_yellow=np.array([40,255,255])

lower_magneta=np.array([140,35,140])
upper_magneta=np.array([160,255,255])

mask_cyan=cv2.inRange(image2,lower_cyan,upper_cyan)
mask_yellow=cv2.inRange(image2,lower_yellow,upper_yellow)
mask_magneta=cv2.inRange(image2,lower_magneta,upper_magneta)

result=cv2.bitwise_and(image,image,mask=mask_cyan)
result=cv2.bitwise_and(result,result,mask=mask_yellow)
result=cv2.bitwise_and(result,result,mask=mask_magneta)

image[mask_cyan>0]=(0,0,255)
image[mask_yellow>0]=(255,0,0)
image[mask_magneta>0]=(0,255,0)

plt.imshow(image,cmap='gray')
```
